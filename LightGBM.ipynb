```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import roc_auc_score

import model_LGBM
import evaluation
#from sklearn.metrics import confusion_matrix
```

# Load and split

```python
data = pd.read_csv('Data/preprocessed_data.csv')
```

```python
data.TRUE_APPETENCE_Epargne_liquide = data.TRUE_APPETENCE_Epargne_liquide.astype(int)
y = pd.DataFrame(data.loc[:,'TRUE_APPETENCE_Epargne_liquide'])
X = data.drop('TRUE_APPETENCE_Epargne_liquide', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train=pd.get_dummies(X_train, drop_first=True)
X_test=pd.get_dummies(X_test, drop_first=True)
```

```python
#np.array(y_train).reshape(1, -1)
```

# Model init and config

```python
model, total_time = model_LGBM.LGBM_model(X_train, y_train, X_test, y_test)
y_pred = model.predict(X_test)
```

# Results and matrices

```python
print('Accuracy score:', accuracy_score(y_test, y_pred))

print("Positive appetance in calibrated prediction: ",np.sum(y_pred)/len(y_pred))
print("Positive appetance in real test set: ", sum(y_test.iloc[:,0].tolist())/len(y_test.iloc[:,0].tolist()))
```

```python
evaluation.plot_confusion_matrix(y_test, y_pred, classes=np.array([0,1]), model_name='LGBM', normalize=True, title='LGBM confusion matrix')
evaluation.plot_confusion_matrix(y_test, y_pred, classes=np.array([0,1]), model_name='LGBM', normalize=False, title='LGBM confusion matrix')
```
